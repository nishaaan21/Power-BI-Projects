```python
df["Acquisition_Cost_clean"] = (
    df["Acquisition_Cost"]
    .str.replace("$", "", regex=False)
    .str.replace(",", "", regex=False)
    .astype(float))
```

```python
df["Date"] = pd.to_datetime(df["Date"])
```

```python
df["Conversion_Rate"] = pd.to_numeric(df["Conversion_Rate"], errors="coerce")
df["ROI"] = pd.to_numeric(df["ROI"], errors="coerce")
df["Clicks"] = pd.to_numeric(df["Clicks"], errors="coerce")
df["Impressions"] = pd.to_numeric(df["Impressions"], errors="coerce")
df["Engagement_Score"] = pd.to_numeric(df["Engagement_Score"], errors="coerce")
```

```python
time_agg = df.groupby("Date").agg(
    avg_conv_rate=("Conversion_Rate", "mean"),
    avg_roi=("ROI", "mean"),
    avg_clicks=("Clicks", "mean")
).reset_index()
```

```python
fig, ax1 = plt.subplots(figsize=(10, 5))
ax2 = ax1.twinx()

sns.lineplot(data=time_agg, x="Date", y="avg_conv_rate", ax=ax1, label="Conversion Rate", color="tab:blue")
sns.lineplot(data=time_agg, x="Date", y="avg_roi", ax=ax1, label="ROI", color="tab:orange")

sns.lineplot(data=time_agg, x="Date", y="avg_clicks", ax=ax2, label="Clicks", color="tab:green")

ax1.set_title("Trend of Conversion Rate, ROI and Clicks Over Time")
ax1.set_xlabel("Date")
ax1.set_ylabel("Rate / ROI")
ax2.set_ylabel("Average Clicks")
lines_1, labels_1 = ax1.get_legend_handles_labels()
lines_2, labels_2 = ax2.get_legend_handles_labels()
ax1.legend(lines_1 + lines_2, labels_1 + labels_2, loc="upper left")

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

sns.histplot(df["Conversion_Rate"], kde=True, ax=axes[0], color="tab:blue")
axes[0].set_title("Distribution of Conversion Rate")
axes[0].set_xlabel("Conversion Rate")
axes[0].set_ylabel("Frequency")

sns.histplot(df["ROI"], kde=True, ax=axes[1], color="tab:orange")
axes[1].set_title("Distribution of ROI")
axes[1].set_xlabel("ROI")
axes[1].set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.barplot(
    data=df,
    x="Campaign_Type",
    y="Conversion_Rate",
    estimator=np.mean,
    ci="sd"
)
plt.title("Average Conversion Rate by Campaign Type")
plt.xlabel("Campaign Type")
plt.ylabel("Average Conversion Rate")
plt.xticks(rotation=30, ha="right")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.boxplot(data=df, x="Channel_Used", y="ROI")
plt.title("ROI Distribution by Channel")
plt.xlabel("Channel")
plt.ylabel("ROI")
plt.xticks(rotation=30, ha="right")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.scatterplot(
data=df,
x="Acquisition_Cost_clean",
y="Conversion_Rate",
hue="Channel_Used",
alpha=0.6
)
sns.regplot(
data=df,
x="Acquisition_Cost_clean",
y="Conversion_Rate",
scatter=False,
color="black"
)
plt.title("Relationship Between Acquisition Cost and Conversion Rate")
plt.xlabel("Acquisition Cost (USD)")
plt.ylabel("Conversion Rate")
plt.tight_layout()
plt.show()
```

```python
channel_agg = df.groupby("Channel_Used").agg(
avg_roi=("ROI", "mean"),
avg_cost=("Acquisition_Cost_clean", "mean")
).reset_index()
fig, ax1 = plt.subplots(figsize=(8, 4))
ax2 = ax1.twinx()
sns.barplot(
data=channel_agg,
x="Channel_Used",
y="avg_roi",
ax=ax1,
color="tab:blue",
alpha=0.7
)
sns.lineplot(
data=channel_agg,
x="Channel_Used",
y="avg_cost",
ax=ax2,
marker="o",
color="tab:red"
)
ax1.set_title("Average ROI and Acquisition Cost by Channel")
ax1.set_xlabel("Channel")
ax1.set_ylabel("Average ROI", color="tab:blue")
ax2.set_ylabel("Average Acquisition Cost", color="tab:red")
ax1.tick_params(axis="x", rotation=30)
plt.tight_layout()
plt.show()
```

```python
aud_agg = df.groupby("Target_Audience", as_index=False)["Conversion_Rate"].mean()
aud_agg = aud_agg.sort_values("Conversion_Rate", ascending=False)

plt.figure(figsize=(8, 4))
sns.barplot(data=aud_agg, x="Target_Audience", y="Conversion_Rate", palette="viridis")
plt.title("Average Conversion Rate by Target Audience")
plt.xlabel("Target Audience")
plt.ylabel("Average Conversion Rate")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
seg_agg = df.groupby("Customer_Segment", as_index=False)["Engagement_Score"].mean()
seg_agg = seg_agg.sort_values("Engagement_Score", ascending=False)

plt.figure(figsize=(8, 4))
sns.barplot(data=seg_agg, x="Customer_Segment", y="Engagement_Score", palette="magma")
plt.title("Engagement Score by Customer Segment")
plt.xlabel("Customer Segment")
plt.ylabel("Average Engagement Score")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.scatterplot(
data=df,
x="Engagement_Score",
y="Conversion_Rate",
hue="Customer_Segment",
alpha=0.6
)
sns.regplot(
data=df,
x="Engagement_Score",
y="Conversion_Rate",
scatter=False,
color="black"
)
plt.title("Engagement vs Conversion Rate")
plt.xlabel("Engagement Score")
plt.ylabel("Conversion Rate")
plt.tight_layout()
plt.show()
```

```python
loc_roi = df.groupby("Location", as_index=False)["ROI"].mean()
loc_roi = loc_roi.sort_values("ROI", ascending=True)
plt.figure(figsize=(8, 6))
sns.barplot(data=loc_roi, x="ROI", y="Location", palette="coolwarm")
plt.title("Average ROI by Location")
plt.xlabel("Average ROI")
plt.ylabel("Location")
plt.tight_layout()
plt.show()
```

```python
lang_conv = df.groupby("Language", as_index=False)["Conversion_Rate"].mean()
lang_conv = lang_conv.sort_values("Conversion_Rate", ascending=False)

plt.figure(figsize=(8, 4))
sns.barplot(data=lang_conv, x="Language", y="Conversion_Rate", palette="cubehelix")
plt.title("Average Conversion Rate by Language")
plt.xlabel("Language")
plt.ylabel("Average Conversion Rate")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
loc_ci = df.groupby("Location").agg(
    total_clicks=("Clicks", "sum"),
    total_impressions=("Impressions", "sum")
).reset_index()

loc_ci = loc_ci.sort_values("total_impressions", ascending=False)

x = np.arange(len(loc_ci["Location"]))
width = 0.35

fig, ax = plt.subplots(figsize=(10, 4))
ax.bar(x - width/2, loc_ci["total_impressions"], width, label="Impressions", color="tab:gray")
ax.bar(x + width/2, loc_ci["total_clicks"], width, label="Clicks", color="tab:blue")

ax.set_title("Impressions and Clicks by Location")
ax.set_xlabel("Location")
ax.set_ylabel("Count")
ax.set_xticks(x)
ax.set_xticklabels(loc_ci["Location"], rotation=45, ha="right")
ax.legend()
plt.tight_layout()
plt.show()
```
